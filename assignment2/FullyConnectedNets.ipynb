## Inline Question 1: 

We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

## Answer:

Both 1 and 2, Sigmoid and ReLU, can lead to bad (zero or close to zero) gradient flow during backpropagation. Leaky ReLu does NOT have this problem.

With Sigmoid, the vanishing gradients problem happens, in the one dimensional case, when the sigmoid input X has a value in either extreme, either a very large positive or very negative value. With either of those extreme input values the output will fall into regions of the sigmoid function where it is flat or nearly flat, resulting in a local gradient of 0 or near zero. Then, when per the chain rule that local gradient is multiplied by the upstream gradient, it  will result in a gradient that is zero or very close to zero flowing down to next layers of backpropagation.

With ReLU, in the one dimensional case, it does not have an issue with killing gradients when the input X is positive. However, when X is 0 or negative it falls respectively in the kink or the flat region of the ReLu function, making the local gradient 0. Since we apply the chain rule in backpropagation, that local gradient of 0 when multiplied by the upstream gradient will result in 0 and will therefore, also by the application of the chain rule downstream, kill the gradient flow to the downstream nodes.

Leaky ReLU, similar to ReLu does not have an issue with gradient flow in the positive regime. Unlike regular ReLu, Leaky ReLu addresses the gradient flow issue in the negative regime by introducing a slight negative slope to the function in that regime in order to avoid killing gradients.

I noticed that in comparison the five layer network was more difficult to train and more sensitive to the initialization scale than the three layer network. 

Deeper networks can be harder to train due to poor gradient flow and collapsing distributions of the activations. Both of these are accentuated as networks get deeper due to the compounding factor of successive matrix multiplications. 

Further explaining, the training of neural networks, both in the forward pass as well as the backward pass, goes through successive matrix multiplications, where the input of a layer depends on the output of the previous layer. That is the weights on one layer will be multiplied by the output of the previous layer to calculate this layer's output and so on. Because of these "chained" multiplications, deeper networks, which have more multiplications chained together, are more susceptible to issues related to activation inputs either going towards zero (in the case when weights are very small) or exploding or falling into saturating regimes (in the case weights are too big).

Since the gradient calculation is dependent on the forward pass values and also relies on similar successive multiplications, the issues described in the previous paragraph can similarly impacts gradients in backpropagation and can lead to very small gradients or exploding gradients.

Another issue, as eluded to in the first paragraph, is that of collapsing distributions which is also accentuated in deeper networks. When using small weights, as training progresses through the layers, more of the activation values will go toward zero, causing the distribution of the activation inputs to shrink and collapse towards zero. This can be remediated by initialization strategies that take into consideration the number of inputs when initializing the weights, in order to keep the variance of the output the same as the variance of the input.

## Inline Question 3:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer: 

AdaGrad can have training issues due to the continuously decaying update step which can be seen in the formula above. AdaGrad keeps a continuously accumulating estimate of square gradients (cache above) which is used in the denominator of the step calculation. As that estimate of the square gradient gets larger and larger, the update step will get smaller and smaller, causing the network to learn slower or potentially even get stuck in a saddle point or local minima.

Adam would not have the same issue. Adam sort of combines the design principles from SGD Momentum, AdaGrad and RMSProp. Like RMSProp, Adam uses a decay rate to calculate a moving average of squared gradients, instead of the continuously accumulating estimate of square gradients seen in AdaGrad, preventing the slow down seen by John when training a network with AdaGrad.
