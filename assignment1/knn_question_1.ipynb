### KNN Question 1 exploration

```python
# Run some setup code for this notebook.

import random
import numpy as np
from cs231n.data_utils import load_CIFAR10
import matplotlib.pyplot as plt

from __future__ import print_function

# This is a bit of magic to make matplotlib figures appear inline in the notebook
# rather than in a new window.
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# Some more magic so that the notebook will reload external python modules;
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
# Load the raw CIFAR-10 data.
cifar10_dir = 'cs231n/datasets/cifar-10-batches-py'
X_train, y_train, X_test, y_test = load_CIFAR10(cifar10_dir)

# As a sanity check, we print out the size of the training and test data.
print('Training data shape: ', X_train.shape)
print('Training labels shape: ', y_train.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
```

```python
# Subsample the data for more efficient code execution in this exercise
num_training = 5000
mask = list(range(num_training))
X_train = X_train[mask]
y_train = y_train[mask]

num_test = 500
mask = list(range(num_test))
X_test = X_test[mask]
y_test = y_test[mask]
```

```python
# Reshape the image data into rows
X_train = np.reshape(X_train, (X_train.shape[0], -1))
X_test = np.reshape(X_test, (X_test.shape[0], -1))
print(X_train.shape, X_test.shape)
```

```python
from cs231n.classifiers import KNearestNeighbor

# Create a kNN classifier instance. 
# Remember that training a kNN classifier is a noop: 
# the Classifier simply remembers the data and does no further processing 
classifier = KNearestNeighbor()
classifier.train(X_train, y_train)
dists = classifier.compute_distances_two_loops(X_test)
print(dists.shape)
```

```python
plt.rcParams['figure.figsize'] = (10,10)
plt.imshow(dists[0:250,0:250], interpolation='none')
plt.show()
```

```python
# Visualize data from rows 200 to 249 where we see some bright rows
plt.rcParams['figure.figsize'] = (20,5)
plt.imshow(dists[200:250,0:250], interpolation='none')
plt.show()
```

```python
# Since the bright rows indicate larger distances, sum each row 
# to pinpoint which ones are showing greater distance overall
dists_sum = np.sum(dists, axis=-1)
print(dists_sum.shape)

# Look at sum of distances specifically for test indexes 200 to 219
start_idx = 200
# Display sorted sums with their respective index
sorted_idxs = np.argsort(dists_sum[200:220]) + start_idx
for i in sorted_idxs:
    print(i, dists_sum[i,])
```

We see that test images with indexes 212, 216, 215 and 218 have much larger distances than the rest. Lets visualize the images to understand the reason.

```python
nrow = 5
ncol = 10
plt.rcParams['figure.figsize'] = (ncol, nrow+2) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

img_idx=200
plt_idx = 0
for i in range(nrow):
    for y in range(ncol):
        plt_idx += 1
        plt.subplot(nrow, ncol, plt_idx)
        visualize = np.reshape(X_test[img_idx], (32, 32, 3))
        plt.imshow(visualize.astype('uint8'))
        plt.axis('off')
        plt.title(img_idx)
        img_idx += 1
plt.show()
```

Looking at the commonality between images 212, 216, 215 and 218, it seems they have much more white pixels than other images. Lets expand the distance comparison to include test images with index 200 to 249 and look at the distances for 238 and 239 which also have a lot of white to further validate this finding.

```python
# Look at sum of distances specifically for test indexes 200 to 249
start_idx = 200
# Display sorted sums with their respective index
sorted_idxs = np.argsort(dists_sum[200:250]) + start_idx
# Print indexes for the 20 images in the range 200:250 with greatest sum of distances
for item in sorted_idxs[:-20:-1]:
    print(item, dists_sum[item])
    
# And also see what are some of the smallest distances
print("\nSmall Sum of Distances in indexes 200 to 249")
for item in sorted_idxs[:10]:
    print(item, dists_sum[item])
```

We see that in fact test images 238 and 239 have large sum of distances. But we also notice that 249 has large sum of distances. Test image 249 has a lot of nearly black pixels. Black RGB is (0,0,0). This points to evidence that while it is more common to have white background and, therefore it is more common to see immages with large sum of distances with a large amount of white pixels, in fact having a lot of pixels in either extreme of RGB would account for large distances.

```python
# Get a sampled look at pixels for image 249
print(np.reshape(X_test[249], (32, 32, 3)))
```

```python
#Finally, out of all 500 test images, get the top 50 largest distance to verify findings.
# Display sorted sums with their respective index
sorted_idxs = np.argsort(dists_sum)
# Print indexes for the 10 images in the range 200:250 with greatest sum of distances
top50 = sorted_idxs[-50:]
print("Images with top 50 sum of distances\n",top50)
print("\n idx  Sum_of_Distances")
for i in top50:
    print("%4d  %.0f" % (i,dists_sum[i]))
```

```python
# Visualize the test images with top 50 largest sum of distances
# out of the whole set of 500
nrow = 5
ncol = 10
plt.rcParams['figure.figsize'] = (ncol, nrow+2) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

sidx = 0
plt_idx = 0
for i in range(nrow):
    for y in range(ncol):
        plt_idx += 1
        plt.subplot(nrow, ncol, plt_idx)
        visualize = np.reshape(X_test[top50[sidx]], (32, 32, 3))
        plt.imshow(visualize.astype('uint8'))
        plt.axis('off')
        plt.title(top50[sidx])
        sidx += 1
plt.show()
```

```python
# Visualize images in index between 200 and 249
# in ascending order of sum of distance
nrow = 5
ncol = 10
plt.rcParams['figure.figsize'] = (ncol, nrow+2) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

sidx = 0
plt_idx = 0
for i in range(nrow):
    for y in range(ncol):
        plt_idx += 1
        plt.subplot(nrow, ncol, plt_idx)
        visualize = np.reshape(X_test[sorted_idxs[sidx]], (32, 32, 3))
        plt.imshow(visualize.astype('uint8'))
        plt.axis('off')
        plt.title(sorted_idxs[sidx])
        sidx += 1
plt.show()
```

```python
# For a given image, get mean pixel value for the image on each of the RGB channels
def spot_check(img_idx):
    print(img_idx, np.mean(np.reshape(X_test[img_idx], (1024, 3)), axis=0), dists_sum[img_idx])
```

```python
# From the images in test in the 200 to 249 index range,
# do spot_check for the two images with lowest sum of dist and the two
# images with highest sum of distance.
print("Images with lowest sum of dist in test index range 200:249 ")
spot_check(sorted_idxs[0])
spot_check(sorted_idxs[1])
print("\nHighest sum of dist")
spot_check(sorted_idxs[-2])
spot_check(sorted_idxs[-1])
```

We can see that the images with lower sum of distances have mean RGB values that are more in the middle of the spectrum (102-128), while the images with higher sum of distances have mean RGB toward the extreme of the RGB spectrum (207-237)

Again visualize the 50 images from the whole test set with highest sum of distances.

```python
# Visualize the test images with top 50 largest sum of distances
# out of the whole set of 500
nrow = 5
ncol = 10
plt.rcParams['figure.figsize'] = (ncol, nrow+2) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

sidx = 0
plt_idx = 0
for i in range(nrow):
    for y in range(ncol):
        plt_idx += 1
        plt.subplot(nrow, ncol, plt_idx)
        visualize = np.reshape(X_test[top50[sidx]], (32, 32, 3))
        plt.imshow(visualize.astype('uint8'))
        plt.axis('off')
        plt.title(top50[sidx])
        sidx += 1
plt.show()
```

```python
# Call spot_check to look at pixels for images 447 and 297,
# the two airplane in blue sky above
# And also image 235 (visualized in the previous section) which is
# also an airplane against blue sky without as much sum of dist
print("\nAirplane imgs with high sum of dist")
spot_check(447)
spot_check(297)
print("\nAirplane imgs with relatively low sum of dist")
spot_check(235)
```

Although these are all airplanes against blue skies, while images 447 and 297 have "bluer blues", resulting in a high mean value for their B channels and also fairly high values of green, image 235 has a mean RGB values that are not as much in the extremes of the spectrum, resulting in a lower sum of distances.

Finally look at some images with many dark pixels.

```python
print("\nVery dark img, 118, with very high sum of dist")
spot_check(118)
print("\nCat img with pixels in extremes of very light and dark and high sum of distance")
spot_check(245)
```

All of this confirms that the distinctly bright rows, which are also the ones with images in the test set with greatest overall distance from the images in the training set, are due to images that have many of its RGB pixels towards extreme ends of the RGB spectrum. Although many white pixels is more commonly seen, many dark pixels or, similarly, very blue pixels will cause the same brightness in the visualization of the distance matrix.
